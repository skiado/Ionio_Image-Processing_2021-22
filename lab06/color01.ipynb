```python
import numpy as np
from PIL import Image
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline

im1 = Image.open('parrot-rgb.jpg')
#im1.show()
a = np.array(im1)
plt.imshow(a)
print(a.shape)
h_r = np.histogram(a[:,:,0],bins=256)
fig1 = plt.figure()
plt.plot(h_r[0]) 
```

```python
h_g = np.histogram(a[:,:,1],bins=256)
fig2 = plt.figure()
plt.plot(h_g[0])
```

```python
h_b = np.histogram(a[:,:,2],bins=256)
fig3 = plt.figure()
plt.plot(h_b[0])
```

```python
# 
b = np.copy(a)
for x in range(a.shape[0]):
    for y in range(a.shape[1]):
        b[x,y,2] = a[x,y,2] * 1.3
        b[x,y,0] = a[x,y,0] * 0.8
        b[x,y,1] = a[x,y,1] * 0.9
        if b[x,y,1] > 255:
            b[x,y,1] = 255
        if b[x,y,2] > 255:
            b[x,y,2] = 255
plt.imshow(b)
```
