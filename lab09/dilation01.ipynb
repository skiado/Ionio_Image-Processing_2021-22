```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

im1 = Image.open('image5.png')
a = np.array(im1)
#se = np.ones((3,3))
se = np.ones((5,5))
plt.imshow(a, cmap='gray', vmin=0, vmax=255)
```

```python
def dilation(ar,se):
    e = False
    for x in range(se.shape[0]):
        for y in range(se.shape[1]):
            if se[x,y] == 1 and ar[x,y] > 0:
                e = True
    return e
```

```python
c = np.zeros((a.shape[0]+4,a.shape[1]+4))
d = np.zeros(a.shape)
c[2:-2,2:-2] = a
for x in range(2,c.shape[0]-2):
    for y in range(2,c.shape[1]-2):
        if dilation(c[x-2:x+3,y-2:y+3],se) :
            d[x-2,y-2] = 255
        else:
            d[x-2,y-2] = 0
plt.imshow(d, cmap='gray', vmin=0, vmax=255)
```

```python
er = d - a
plt.imshow(er, cmap='gray', vmin=0, vmax=255)
er = er.astype(np.uint8)
im2 = Image.fromarray(er)
im2.save('image5_2orio3.png')
```
